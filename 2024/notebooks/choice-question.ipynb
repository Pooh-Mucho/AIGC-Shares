```python
import lxw
```

### 1、选择，3选1

```python
client = lxw.new_client()
messages = [
    {
        "role": "user",
        "content": """
你是一名小红书的典型年轻女性用户。选出下列租房文案标题中，对你吸引力最强的一个：
'''
A:💫潮流青年的租房首选！🏠现代公寓超in生活，地铁旁边，咖啡馆式阳台等你check in✨
B:💫生活不将就，#心动新公寓#来啦！安静社区🌿，设计师风格🎨，你的潮流栖息地～
C:桃花源里的INS风公寓🌸，青春租房不将就！顶级社区，解锁你的潮流新生活🔑
'''
先说明选择过程与理由，再给出选择结果。
        """
    }
]
response = client.chat.completions.create(
    model = lxw.gpt4_turbo,
    messages = messages,
    n=3
)
lxw.print_response(response)
```
