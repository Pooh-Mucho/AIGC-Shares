#### 2. 宣传语生成

```python
client = lxw.new_client()
messages = [
    {
        "role": "user",
        "content": """
## Role：
你是一名活动策划专家
## Skills:
1. 熟悉不同用户群体的行为特点分析
2. 拥有创意思维，文案可以打动用户
## Background：
龙湖北苑天街是一家著名的购物中心。龙湖北苑天街在3月份，发起一次“循环共生季”的活动，用户可以参与旧衣回收，提供旧衣物，换取限量版的可回收的环保袋。
## Workflow:
1. 分析可能参与本次活动的用户群体，及其行为特点
2. 根据步骤1的分析，结合本次活动内容，设计本次活动的'一句话宣传语'
## Constrains:
1. 活动的'一句话宣传语'描述在25个字以内
2. 确保活动的'一句话宣传语'，对于可能参与活动的用户群体有较高的吸引度
3. 确保活动的'一句话宣传语'，能让用户清晰了解本次活动的内容，包括用户需要提供旧衣服，换取限量版的环保袋，该环保袋是可回收的
## Task：
设计3条本次活动的'一句话宣传语'
## Example:
1. 参与旧衣回收，赢限量版可再生环保袋
## Output：
1. 可能参与本次活动的用户群体特点与行为的深入分析
2. 针对这些用户群体，设计活动的'一句话宣传语'，以XML标签形式提供
<活动描述1> 本次活动的一句话描述 </活动描述1>
<活动描述2> 本次活动的一句话描述 </活动描述2>
<活动描述3> 本次活动的一句话描述 </活动描述3>
        """
    }
]
response = client.chat.completions.create(
    model = lxw.gpt4_turbo,
    messages = messages,
    temperature = 1.1,
    n=10
)
```

```python
def parse_response(response):
    search_pattern = "<活动描述\d+>(.*?)</活动描述\d+>"
    replace_pattern = "[\" ！！。《》“”]"
    result = []
    for choice in response.choices:
        content = choice.message.content
        matches = re.findall(search_pattern, content, re.DOTALL)
        matches = [re.sub(replace_pattern, '', x.strip()) for x in matches]
        result += matches
    return result
```

```python
data = [[x, 0, 0, 0] for x in parse_response(response)]
df = pd.DataFrame(data, index=range(1, len(data)+1), columns=['text', 'times', 'total', 'average'])
df
```

#### 3. 自动评估

```python
def eval_scores(rows):
    prompt = """
## Role
你是一名活动策划专家，负责对活动文案进行评分。

## Profile
1. 你熟悉不同用户群体的行为特点分析
2. 你通过用户群体洞察，对文案进行量化评估

## Background
龙湖北苑天街是一家著名的购物中心。龙湖北苑天街在3月份，发起一次“循环共生季”的活动，用户可以参与旧衣回收，获得限量版的环保袋。

## Workflow
你需要按照评估标准对本次活动的'一句话宣传语'进行评分。本次活动的目标用户如下：
- 环保意识较强的用户：对环境保护有责任感，愿意通过参与旧衣回收等行动来减少资源浪费。
- 对限量版产品感兴趣的用户：逐求独特性，并愿意通过环保行为获得限量版环保袋。
- 时尚意识较强的年轻消费群体：追求潮流，同时关注可持续时尚和环保理念。
- 社会责任感强的居民：希望自己的行为能对社会产生积极影响，不仅仅是消费行为。
- 省钱及实用主义者：看重活动给予的实际利益，注重环保袋的使用价值。
- 家庭主妇：经常处理家庭中的旧物，对旧衣回收的活动感兴趣。

评分标准如下：
- 准确性：1-10分。'一句话宣传语'必须准确传达活动内容，包括用户需要提供旧衣服，换取限量版的环保袋，该环保袋是可回收的。
- 易懂性：1-10分。'一句话宣传语'语言简单、通俗，确保所有层次的用户都能理解。
- 吸引力：1-10分。'一句话宣传语'具有足够的吸引力，激发目标用户群体的兴趣。
- 相关性：1-10分。'一句话宣传语'能够明确与环保、循环利用等概念关联起来。
- 易记性：1-10分。'一句话宣传语'简洁、醒目，便于顾客记忆。
- 独特性：1-10分。'一句话宣传语'具有一定的创意，能够在众多活动中脱颖而出。
- 行动倡导：1-10分。'一句话宣传语'具有一定的号召力，鼓励用户参与到活动中来。
- 情感共鸣:：1-10分。'一句话宣传语'能够触动用户的环保情感，激发他们参与活动的热情。

你通过分析洞察用户，按照评分标准，对'一句话宣传语'进行评分。

## Input
输入为json格式，示例如下：
{"id": 1, "text": "一句话宣传语的内容"}
{"id": 2, "text": "一句话宣传语的内容"}
{"id": 3, "text": "一句话宣传语的内容"}
 ...

## Output
- 基于目标用户和评分原则，对于文案的深入分析和评估
- 该文案的最终得分：
{"score": 80, "id": 1, "text": "一句话宣传语的内容"}
- 基于目标用户和评分原则，对于文案的深入分析和评估
- 该文案的最终得分：
{"score": 79, "id": 2, "text": "一句话宣传语的内容", }
...

## Task
你需要评估的'一句话宣传语'如下：
{TEXT}
"""
    json_text = ''
    for index, row in rows.iterrows():
        json_text += '{' + """ "id": {0}, "text": "{1}" """.format(index, row['text']) + '},\r\n'
    messages = [
        {
            "role": "user",
            "content": prompt.replace('{TEXT}', json_text)
        }
    ]
    response = client.chat.completions.create(
        #model = 'gpt-35-turbo-1106', 
        model = lxw.gpt4_turbo,
        messages = messages,
        top_p = 0.75,
        n=1
    )
    content = response.choices[0].message.content
    matches = re.findall("({.*?score.*?(\\d+).*?id.*?(\\d+).*?text.*?})", content, re.MULTILINE)
    scores = {}
    for match in matches:
        scores[int(match[2])] = int(match[1])
    return content, matches, scores
```

```python
LOOPS = 1
for i in range(LOOPS):
    samples = df.sample(10)
    content, matches, scores = eval_scores(samples)
    df.loc[scores.keys(), 'times'] += 1
    df.loc[scores.keys(), 'total'] += list(scores.values())
```
